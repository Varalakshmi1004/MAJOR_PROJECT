# CREATING THE DATAFRAMES 

```python
import pandas as pd
books=pd.read_excel('mldataset.xls')
users=pd.read_csv('booksusers.csv')
users=users.dropna()
users.reset_index(drop=True)
```

# DATA VISUALIZATION

```python
import matplotlib.pyplot as plt
plt.rc("font",size=15)
users.groupby('review rate').size().plot(kind='bar',color='maroon')
plt.title('RATING DISTRIBUTION')
plt.xlabel('rating')
plt.ylabel('count')
plt.savefig("plot.png",bbox_inches='tight')
plt.show()
```

# MERGING THE TWO DATASETS BASED OF ISBN

```python
combine_book_rating=pd.merge(books,users,on='ISBN')
combine_book_rating
```

# DROPPING UNNECESSARY COLUMNS

```python
columns=['PUBLISHER','YEAR OF PUBLISHMENT','AUTHOR']
combine_book_rating=combine_book_rating.drop(columns,axis=1)
combine_book_rating
```

# DROPPING THE NULL VALUES

```python
combine_book_rating=combine_book_rating.dropna(axis=0,subset=['BOOK NAME'])
combine_book_rating
```

```python
total_ratings=pd.DataFrame(books['ISBN'])
total_ratings=total_ratings.join(books['BOOK NAME'])
total_ratings=total_ratings.join(books['RATING COUNTS'])
total_ratings.head()
```

```python
users_with_rating_counts=combine_book_rating.merge(total_ratings,left_on='BOOK NAME',right_on='BOOK NAME')
cols=['ISBN_y','RATING COUNTS_y','review rate']
users_with_rating_counts=users_with_rating_counts.drop(cols,axis=1)
users_with_rating_counts.rename(columns={'ISBN_x':'ISBN','RATING COUNTS_x':'RATING COUNTS'},inplace=True)
users_with_rating_counts=users_with_rating_counts.dropna()
```

# CREATING A CSR MATRIX

```python
from scipy.sparse import csr_matrix
users_with_rating_counts=users_with_rating_counts.drop_duplicates(['Customer Name','BOOK NAME'])
user_rating_pivot=users_with_rating_counts.pivot(index='BOOK NAME',columns='Customer Name',values='B_NO')
user_rating_pivot.fillna(0,inplace=True)
user_rating_pivot
user_rating_matrix=csr_matrix(user_rating_pivot.values)
```

# IMPORTING KNN TO USE COLLABRATIVE-FILTERING

```python
from sklearn.neighbors import NearestNeighbors
model_knn=NearestNeighbors(metric='cosine',algorithm='brute')
model_knn.fit(user_rating_matrix)
```

```python
import numpy as np
query_index=np.random.choice(user_rating_pivot.shape[0])
print(query_index)
distances,indices=model_knn.kneighbors(user_rating_pivot.iloc[query_index,:].values.reshape(1,-1),n_neighbors=6)
```

```python
user_rating_pivot.index[query_index]
```

```python
for i in range(0,len(distances.flatten())):
    if i==0:
        print('Recommendation of {0}:\n'.format(user_rating_pivot.index[query_index]))
    else:
        print("{0}:{1},with distance of {2}:".format(i,user_rating_pivot.index[indices.flatten()[i]],distances.flatten()[i]))
```
